```python
import pandas as pd 
import numpy as np
from sklearn.impute import SimpleImputer

df = pd.read_csv(r'C:\Users\micky\Downloads\house-votes-84.data.csv')
df.columns = ['party', 'infants', 'water', 'budget', 'physician', 'salvador',
       'religious', 'satellite', 'aid', 'missile', 'immigration', 'synfuels',
       'education', 'superfund', 'crime', 'duty_free_exports', 'eaa_rsa']
df.replace('y',1, inplace = True)
df.replace('n',0, inplace = True)
```

```python
df.head()
```

```python
#開始使用pipeline來完成整套簡單的資料清整與模型套用
from sklearn.pipeline import Pipeline
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report

X = df.drop('party', axis = 1).values
y = df['party'].values

simple_imputer = SimpleImputer(missing_values = '?', strategy = 'most_frequent')
steps  = [('imputation', simple_imputer),
         ('SVM', SVC())]
pipeline = Pipeline(steps)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)

pipeline.fit(X_train, y_train)

pred = pipeline.predict(X_test)

print(classification_report(y_test, pred))

#如果資料的分佈差異過大可先做標準化
"""
from sklearn.preprocessing import StandardScaler
steps  = [('imputation', simple_imputer),
         ('scale', StandardScaler()),
         ('SVM', SVC())]
pipeline.fit(X_train, y_train)
pred = pipeline.predict(X_test)
print(classification_report(y_test, pred))
"""
```
